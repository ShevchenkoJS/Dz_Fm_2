**Задача 1.** Есть инвестиционный проект с денежными потоками по кварталам:
Квартал 0 1 2 3 4 5
денежный поток -1200 100 200 300 400 500
Необходимо принять решение, инвестируем в проект или нет, если ставка дисконтирования 15% годовых.

Для принятия решения оценим чистую приведенную стоимость (NPV) проекта с учетом ставки дисконтирования 15% годовых. Нулевой квартал будет использован только для инвестиции, поэтому дисконтироваться не будет. Для оценки NPV воспользуемся следующей формулой:

\[NPV = ∑(Cash Flow / (1 + r)^t)\]

где NPV - чистая приведенная стоимость, Cash Flow - денежный поток в заданный квартал, r - ставка дисконтирования, t - номер квартала.

Используем библиотеку matplotlib для построения графика.

```python
import matplotlib.pyplot as plt

cash_flows = [-1200, 100, 200, 300, 400, 500]
discount_rate = 0.15
npv = 0

npv_values = []

for t, cash_flow in enumerate(cash_flows):
    npv += cash_flow / (1 + discount_rate) ** t
    npv_values.append(npv)

plt.plot(npv_values)
plt.xlabel('Квартал')
plt.ylabel('Чистая приведенная стоимость')
plt.title('График NPV')
plt.show()

print(f"Чистая приведенная стоимость проекта: {npv:.2f}")

```

*Вывод: NPV отрицательное, проект не будет приносить необходимую прибыль, следовательно инвестиция в проект не рекомендуется.*

**Задача 4.** Есть два инвестиционных проекта со следующими денежными потоками:
Год 0 1 2 3 4 5
Проект А -1000 100 250 450 500 550
Проект Б -1000 200 300 400 450 500
Если стоимость денег равна 10%, и инвестор хочет получить максимальную доходность на вложенный рубль инвестиций, то какой проект он должен выбрать?

Чтобы определить, какой проект будет наиболее доходным при стоимости денег в 10%, необходимо рассчитать чистую текущую стоимость (NPV) для каждого проекта.

Формула NPV для каждого года:
NPV = (CF) / (1+r)^t

Где:
NPV - чистая текущая стоимость
CF - денежный поток в данном году
r - стоимость денег
t - год

Для проекта А:
NPV = (-1000 / (1+0.1)^0) + (100 / (1+0.1)^1) + (250 / (1+0.1)^2) + (450 / (1+0.1)^3) + (500 / (1+0.1)^4) + (550 / (1+0.1)^5)

Для проекта Б:
NPV = (-1000 / (1+0.1)^0) + (200 / (1+0.1)^1) + (300 / (1+0.1)^2) + (400 / (1+0.1)^3) + (450 / (1+0.1)^4) + (500 / (1+0.1)^5)

```python
import matplotlib.pyplot as plt

r = 0.1

cash_flows_A = [-1000, 100, 250, 450, 500, 550]
cash_flows_B = [-1000, 200, 300, 400, 450, 500]

npv_A = 0
npv_B = 0

for i in range(len(cash_flows_A)):
    npv_A += cash_flows_A[i] / (1+r)**i
    npv_B += cash_flows_B[i] / (1+r)**i

print("NPV для проекта А:", npv_A)
print("NPV для проекта Б:", npv_B)

# Построение графика
years = [0, 1, 2, 3, 4, 5]

plt.plot(years, cash_flows_A, label='Проект А')
plt.plot(years, cash_flows_B, label='Проект Б')

plt.xlabel('Год')
plt.ylabel('Денежные потоки')

plt.legend()
plt.grid(True)

plt.show()
```

При стоимости денег в 10%, NPV для проекта А равна 318.63, а NPV для проекта Б равна 348.09. Следовательно, инвестор должен выбрать проект Б, так как он имеет более высокую NPV и приносит большую доходность на вложенный рубль инвестиций.

На графике показаны денежные потоки для каждого проекта по годам.

**Задача 5.** Есть два инвестиционных проекта со следующими денежными потоками:
Год 0 1 2 3 4 5
Проект А -1000 100 250 450 500 550
Проект Б -1000 200 300 400 450 500
Инвестор хочет выбрать один из проектов по критерию ликвидности с учетом временной стоимости денег в размере 10% годовых. Какой он должен выбрать?

Рассчитаем накопленную сумму дисконта для каждого проекта с учетом временной стоимости денег. Для этого воспользуемся формулой:

\[DS = \sum \frac{CF}{(1 + r)^t}\]

Где:
DS - дисконтированная накопленная сумма
CF - денежный поток в году \(t\)
r - ставка дисконтирования
t - год

```python

import matplotlib.pyplot as plt

# Денежные потоки для проекта А
cash_flows_a = [-1000, 100, 250, 450, 500, 550]

# Денежные потоки для проекта Б
cash_flows_b = [-1000, 200, 300, 400, 450, 500]

# Процентная ставка
interest_rate = 0.1

# Расчет дисконтированных значений для каждого года
discounted_values_a = [cash_flow / (1 + interest_rate)**year for year, cash_flow in enumerate(cash_flows_a)]
discounted_values_b = [cash_flow / (1 + interest_rate)**year for year, cash_flow in enumerate(cash_flows_b)]

# Расчет накопленной суммы дисконта
accumulated_discount_a = [sum(discounted_values_a[:year+1]) for year in range(len(discounted_values_a))]
accumulated_discount_b = [sum(discounted_values_b[:year+1]) for year in range(len(discounted_values_b))]

# Построение графика накопленной суммы дисконта
years = range(len(cash_flows_a))
plt.plot(years, accumulated_discount_a, label='Проект А')
plt.plot(years, accumulated_discount_b, label='Проект Б')

plt.xlabel('Годы')
plt.ylabel('Накопленная сумма дисконта')
plt.legend()
plt.show()

```

*Вывод: На графике выведены две кривые - одна для проекта A и другая для проекта Б. Кривая, расположенная выше, представляет проект с наибольшей накопленной суммой дисконта, следовательно, инвестор должен выбрать проект Б.*
